Найти площадь равнбедренной трапеции, у которой основания равны 6 и 10 см, а диагонали взаимно
перпендикулярны.

![Task-8-16](Task-8-16.png)

```python

from sympy import Eq
from sympy import solve
from sympy import symbols
from sympy import sqrt

x = symbols("x", real=True)
AD = 10; BC = 6; AI = x; ID = AI
eq1 = Eq(AD ** 2, AI ** 2 + ID ** 2)
r1 = solve(eq1, AI)
BI = x; CI = BI
eq2 = Eq(BC ** 2,BI ** 2 + CI ** 2)
r2 = solve(eq2, BI);
AI = r1[1]; AF = AD / 2;
IF = sqrt(AI ** 2 - AF ** 2)
BI = r2[1]; BH = BC / 2
IH = sqrt(BI ** 2 - BH ** 2)
HF = IH + IF
S = (BC+AD) * HF / 2
print(eq1.lhs, "=", eq1.rhs)
print(r1)
print(eq2.lhs, "=", eq2.rhs)
print(r2)
print("S =", S)
```
